## t5-LARGE

NEWSROOM

```python
# Step 1: Install and import necessary libraries
!pip install transformers datasets rouge-score nltk --quiet

import torch
from transformers import T5Tokenizer, T5ForConditionalGeneration
from datasets import load_dataset
from rouge_score import rouge_scorer
import nltk
nltk.download("punkt")

# Step 2: Load the T5-Large model and tokenizer
model_name = "t5-large"
tokenizer = T5Tokenizer.from_pretrained(model_name)
model = T5ForConditionalGeneration.from_pretrained(model_name).to("cuda" if torch.cuda.is_available() else "cpu")
device = model.device
print("Using device:", device)

# Step 3: Load Newsroom dataset (from preprocessed files)
newsroom = load_dataset(
    "json",
    data_files={
        "test": "/root/tensorflow_datasets/downloads/manual/newsroom/test.jsonl"
    }
)

# Use a small sample for testing
newsroom_small = newsroom["test"].shuffle(seed=42).select(range(5))

# Step 4: Define summarization function
def summarize(example):
    input_text = "summarize: " + example["text"]
    inputs = tokenizer(input_text, return_tensors="pt", padding="longest", truncation=True, max_length=512).to(device)
    with torch.no_grad():
        outputs = model.generate(inputs.input_ids, max_length=60, num_beams=4)
    return {"generated_summary": tokenizer.decode(outputs[0], skip_special_tokens=True)}

# Step 5: Generate summaries
newsroom_result = newsroom_small.map(summarize)

# Step 6: Compute ROUGE scores
scorer = rouge_scorer.RougeScorer(["rouge1", "rouge2", "rougeL"], use_stemmer=True)

def compute_rouge(example):
    scores = scorer.score(example["summary"], example["generated_summary"])
    return {
        "rouge1": scores["rouge1"].fmeasure,
        "rouge2": scores["rouge2"].fmeasure,
        "rougeL": scores["rougeL"].fmeasure,
    }

newsroom_result = newsroom_result.map(compute_rouge)

# Step 7: Print outputs
for i, ex in enumerate(newsroom_result):
    print(f"\n📰 Sample {i+1}")
    print("Original Summary:", ex["summary"])
    print("Generated Summary:", ex["generated_summary"])
    print("ROUGE-1:", round(ex["rouge1"], 4), "| ROUGE-2:", round(ex["rouge2"], 4), "| ROUGE-L:", round(ex["rougeL"], 4))

# Step 8: Accuracy metric
accuracy = sum([ex["rougeL"] > 0.5 for ex in newsroom_result]) / len(newsroom_result)
print(f"\n✅ Approximate Accuracy (ROUGE-L > 0.5): {round(accuracy * 100, 2)}%")
```

```python
import matplotlib.pyplot as plt

# ✅ Step 1: Extract ROUGE scores
rouge1_scores = [ex["rouge1"] for ex in newsroom_result]
rouge2_scores = [ex["rouge2"] for ex in newsroom_result]
rougeL_scores = [ex["rougeL"] for ex in newsroom_result]

# ✅ Step 2: Plot
plt.figure(figsize=(10, 6))
plt.plot(rouge1_scores, label="ROUGE-1", marker='o')
plt.plot(rouge2_scores, label="ROUGE-2", marker='s')
plt.plot(rougeL_scores, label="ROUGE-L", marker='^')

plt.title("T5-Large on Newsroom: ROUGE Scores for 5 Samples")
plt.xlabel("Sample #")
plt.ylabel("ROUGE F1 Score")
plt.ylim(0, 1)
plt.grid(True)
plt.legend()
plt.show()
```
